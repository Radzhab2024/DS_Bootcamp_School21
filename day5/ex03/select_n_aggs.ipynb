```python
import pandas as pd
```

## 1. Загрузка созданного раннее .json файла в фрейм данных.

```python
df = pd.read_json('../ex02/auto.json',
                  orient= 'records')
df = df.set_index('CarNumber')
df.tail()
```

## 2. Выполняем следующие действия:

```python
df_filtered = df[df['Fines'] > 2100]
df_filtered.tail()
```

```python
df_filtered = df[(df['Fines'] > 2100) & (df['Refund'] == 2)]
df_filtered
```

```python
df_filtered = df[df['Model'].isin(['Corolla', 'Focus'])]
df_filtered
```

```python
numbers_list = ['Y7689C197RUS', '92928M178RUS', '7788KT197RUS', 'H115YO163RUS', 'X758HY197RUS']
df_filtered = df[df.index.isin(numbers_list)]
df_filtered
```

## 3. Проводим агрегации с помощью make.

```python
median_fines_by_make = df.groupby('Make')['Fines'].median()
median_fines_by_make
```

```python
median_fines_by_make = df.groupby(['Make', 'Model'])['Fines'].median()
median_fines_by_make
```

```python
count_fines_by_make = df.groupby(['Make', 'Model'])['Fines'].count()
count_fines_by_make
```

```python
min_max_fines_by_make = df.groupby(['Make', 'Model'])['Fines'].agg(['min', 'max'])
min_max_fines_by_make
```

```python
std_fines_by_make = df.groupby(['Make', 'Model'])['Fines'].std()
std_fines_by_make
```

## 4. Проводим агрегации с помощью CarNumber.

```python
violations_count = df.groupby(level=0)['Fines'].count().sort_values(ascending=False)
violations_count.head()
```

```python
top_violation_car = violations_count.idxmax()
top_violation_data = df.loc[top_violation_car]
top_violation_data
```

```python
violations_sum = df.groupby(level=0)['Fines'].sum().sort_values(ascending=False)
violations_sum
```

```python
top_payer_car = violations_sum.idxmax()
top_payer_data = df.loc[top_payer_car]
top_payer_data
```

```python
unique_models = df.groupby(level=0)['Model'].nunique().sort_values(ascending=False)
unique_models.head()
```
